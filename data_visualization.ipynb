```python
import seaborn as sns
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/teachingratings.csv'
df = pd.read_csv(df_url)
```

```python
division_eval = df.groupby('division')['eval'].mean().reset_index()
```

```python
sns.set(style = "dark")
ax = sns.barplot(x = "division" , y = "eval", data = division_eval)
```

bar graph is used the average

```python
ax = sns.scatterplot(x = 'age', y = 'eval', data = df)
```

scatterplot is used to find the relationship between the age and the teaching evaluation score. In this scatterplot, we can see that there is no patter or trend and this kind of graph is called as no trend. This is because the data that have been used is random. 

```python
ax = sns.scatterplot(x = 'age', y = 'eval', hue = "gender", data = df )
```

hue is assigns different colors or visual markers to the data points based on the unique categories of the specified variable. This helps to create visual distinctions and provides additional information about the data.



```python
ax = sns.boxplot(x = 'credits', y = 'beauty', data = df)
```

```python
sns.catplot(x = "gender" , kind = "count", data = df)
#in catplot kind is used instead of y to represent y axis. 
```

```python
sns.catplot(x = "gender" , hue= 'tenure' , kind = "count", data = df)
```

this shows that gender (male) and tenure is connected and it shows that gender affects in the number of class taught by the professor.  

```python
sns.catplot(x = "gender", hue = "tenure", row = "division", kind = "count", data = df, height = 2, aspect = 2)
```

```python
sns.relplot(x = "age", y = "eval", hue = "gender", row = "tenure", data = df , height = 4, aspect = 2)
```

```python
ax =sns.distplot(df[df['native'] == 'yes']['beauty'], color = 'orange', kde = False)
ax =sns.distplot(df[df['native'] == 'no']['beauty'], color = 'blue', kde = False)
plt.show()
```

```python
sns.set(style ="dark")
ax = sns.barplot(x = "minority" , y = "age", data = df)
```

this graph shows the instructor who is not a minority has a high age in compare to the people who are from the minority.
