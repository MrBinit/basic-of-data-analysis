```python
import seaborn as sns
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/teachingratings.csv'
df = pd.read_csv(df_url)
```

```python
df.prof.nunique()
```

```python
df['age'].mean()
```

```python
df['age'].std()
```

```python
duplicates_df = df.drop_duplicates(subset = ['prof'])
duplicates_df.head(19)
#drop_duplicate is used to delete the duplicate rows. 
```

```python
duplicates_df['age'].mean()
```

```python
duplicates_df['age'].std()
```

```python
division_eval = df.groupby('division')['eval'].mean().reset_index()
```

```python
sns.set(style = "dark")
ax = sns.barplot(x = "division" , y = "eval", data = division_eval)
```

bar graph is used the average

```python
ax = sns.scatterplot(x = 'age', y = 'eval', data = df)
```

scatterplot is used to find the relationship between the age and the teaching evaluation score.
